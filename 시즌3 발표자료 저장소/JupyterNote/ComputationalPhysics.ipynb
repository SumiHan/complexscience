* Newton's method

```python

def root_newton(f, df, guess, tolerance = 1.0e-6):
    """
    Uses Newton's method to find a value x near "guess"
    for which f(x) = 0, within the tolerance given.
    Default tolerance is 1.03-6, if no tolerance is 
    specified in the function call.
    
    It is require to pass this function both f(s) and f'(x).
    """
    x = guess
    dx = 2 * tolerance          # initial dx > delta
    while dx > tolerance:       # main loop -- loop until
                                # dx < tolerance
        x1 = x - f(x)/df(x)     # Point-slop form of line
        dx = abs(x - x1)        # how much have things changed?
        x = x1
        print("dx: " + str(round(dx, 3))
              + " x: " + str(round(x, 3)) \
              + " fx: " + str(round(f(x), 3)) \
              + " dfx: " + str(round(df(x), 3)))
    return x                    # Best value so far


from math import *

theta_0 = root_newton(sin, cos, 0.8)
print( round(theta_0, 3))
```

```python
import os, sys, string, math

## for Graph
import numpy as np
import matplotlib.pyplot as plt

import bokeh.io
import bokeh.mpl
from bokeh.plotting import figure, output_file, show

%matplotlib inline


# Set up Bokeh for inline viewing
bokeh.io.output_notebook()

## Logic
def integrate(f_prime, a, b, m):
    dx = (b - a) / m
    s = 0.
    for i in range(0, m):
        s = s + f_prime(a + i * dx) * dx
    return s

func = lambda x: x**3
start = 0
end = 2
step = 10

print (integrate(func, start, end, step))

## for Graph
# 시작점과 끝점이 있으므로 원래 보다 1개 더 많은 스텝이 필요함에 주의
xVals = np.linspace(start, end, step + 1)
yVals = list(map(func, xVals))

p = figure(
    tools="pan,box_zoom,reset,save",
    y_range=[0,func(end)], title="Integration x^3",
    x_axis_label="x", y_axis_label="x^3"
    )

p.line(xVals, yVals, legend="y = x^3")
p.quad(bottom=0,left=xVals[:-1], right=xVals[1:], top=yVals, alpha=0.5)
show(p)
```

```python
# 오른쪽 값을 사용한 면적 방식으로 그려보기
p = figure(
    tools="pan,box_zoom,reset,save",
    y_range=[0,func(end)], title="Integration x^3",
    x_axis_label="x", y_axis_label="x^3"
    )

p.line(xVals, yVals, legend="y = x^3")
p.quad(bottom=0,left=xVals[:-1], right=xVals[1:], top=yVals[1:], alpha=0.5)
show(p)
```

# Differentiation

Taylor expansion

```python
def derivate(f, x, dx):
    return (f(x + dx) - f(x)/dx)

print (derivate(lambda x: x**2, 0., 0.000001))
```
